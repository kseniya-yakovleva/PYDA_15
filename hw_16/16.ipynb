2. Выбросы. В столбце age выбросом будет максимальное значение 9, т.к. предполагаются значения 1 и 2. В столбце pulse максимальное значение тоже может быть выбросом (хотя верхний порог в описании не указывается). Выбросы по возрасту можно заменить на 2, т.к. возможно это неправильная запись (в значениях только 1 и 9)

```python
q1 = df.age.quantile(0.25)
q3 = df.age.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df.age.between(lower_bound, upper_bound, inclusive=True)]
remove_outliers
```

```python
pd.concat([df, remove_outliers]).drop_duplicates(keep=False) #кол-во выбросов по возрасту, 24 значения
```

```python
#заменить значения 9 на 2
```

```python
df.age.replace('9', '2', inplace=True)
```

```python
df.describe()
```

```python
df.dropna(thresh=7).info()
```

```python
df.groupby(['peristalsis', 'abdominal_distension']).abdomen.median()
```

```python
df.abdomen.fillna(df.groupby(['peristalsis', 'abdominal_distension']).abdomen.transform('median'), inplace=True)
```
