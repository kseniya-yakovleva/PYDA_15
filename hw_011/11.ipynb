* Задание 2 По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
coutries_filter = power[(power['country'] == 'Latvia')|(power['country'] == 'Lithuania')|(power['country'] == 'Estonia')]
coutries_filter.head() # фильтр на страны
```

```python
coutries_filter[(coutries_filter.quantity > 0)].head() #значения потребления больше 0
```

```python
filter_year = coutries_filter[(coutries_filter.year > 2005)|(coutries_filter.year < 2010)]
filter_year.head() #фильтр на период 2005-2010
```

```python
category_filter = filter_year[(filter_year.category == 4)|(filter_year.category == 12)|(filter_year.category == 21)]
category_filter.quantity.sum() # фильтр по категориям и сумма из значений
```
